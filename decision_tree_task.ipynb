```python
# Libraries we'll use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# From scikit-learn for the model
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load dataset
iris = load_iris()

# Features (like petal size, sepal width, etc.)
X = iris.data

# Labels (which flower it is: 0=setosa, 1=versicolor, 2=virginica)
y = iris.target

# See the features
print("Feature Names:", iris.feature_names)
print("Target Names:", iris.target_names)
```

```python
# Split the data: 70% for training, 30% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Create the model
model = DecisionTreeClassifier()

# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = model.predict(X_test)

# Accuracy check
print("Accuracy:", accuracy_score(y_test, y_pred))

# Detailed report
print(classification_report(y_test, y_pred))
```

```python
# Draw the tree
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree (Iris Dataset)")
plt.show()
```

This decision tree model was built using the Iris dataset to classify flowers based on their measurements. The model mainly splits based on petal length and sepal width.

It achieved around 95% accuracy, showing good performance. The tree visualization is simple and clearly shows how decisions are made to classify each flower type.
